```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cards_data.csv')
df.head()
```

## Correlation between power and toughness


```python
df['power'] = pd.to_numeric(df['power'], errors='coerce')
df['toughness'] = pd.to_numeric(df['toughness'], errors='coerce')


pt_only = df[(df['power'].notnull()) & (df['toughness'].notnull())]
pt_only = pt_only[['power', 'toughness']]
sns.regplot(data=pt_only, x='power', y='toughness')
```
