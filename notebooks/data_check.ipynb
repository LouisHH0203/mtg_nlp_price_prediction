## Find out which colunms to keep / cleaning nans


```python
df.columns
```

```python
df = df[['name', 'colorIdentity', 'colors', 'type', 'types',  'subtypes', 'supertypes', 'text', 'keywords', 'power', 'toughness' ,
          'manaValue', 'manaCost', 'loyalty', 'rarity', 'setCode', 'uuid', ]]
df.head(5)
```

```python
df.drop_duplicates(subset='name', inplace=True)
df.head()
```

```python
df['supertypes'].value_counts()
```

```python
df['supertypes'].fillna('No Types', inplace=True)
df['power'].fillna('Not Creature', inplace=True)
df['toughness'].fillna('Not Creature', inplace=True)
df['subtypes'].fillna('Not Creature', inplace=True)
df['keywords'].fillna('No Keywords', inplace=True)
df['loyalty'].fillna('Not Planeswalker', inplace=True)
df.head()
```

```python
print(f'There are currently {df["name"].nunique()} cards')
print(f'There has been {df["setCode"].nunique()} sets printed')
```
