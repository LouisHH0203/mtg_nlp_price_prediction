```python
data = call()
data.keys()
card_data = data['data']
card_data
```

```python
keys =[
    'cmc',                  # Converted mana cost
    'mana_cost',            # Pips and cost
    'colors',               # Colors associated with the card
    'type_line',            # Type of the card
    'oracle_text',          # Rules text of the card
    'textless',             # Vanilla cards
    'keywords',             # Special abilities or keywords
    'rarity',               # Rarity of the card
    'reserved',             # Whether the card is on the Reserved List
    'foil', 'nonfoil',      # Whether the card is available in foil or non-foil
    'set',                  # Set the card belongs to
    'released_at',          # Release date of the card
    'booster',              # Whether the card is available in booster packs
    'prices',               # Information about the card's prices
    'name',                 # Name of the card
    'card_back_id',         # ID associated with the card's back
]

card_list =[]
for cards in card_data:
    filtered_dict = {key: cards.get(key) for key in keys if key in cards}
    card_list.append(filtered_dict)

card_list
```

```python
df = pd.DataFrame(card_list)
df
```
